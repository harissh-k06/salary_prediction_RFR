```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('final salary.csv')
```

```python
dataset
```

```python

num_unique_job_titles = dataset['Job Title'].nunique()
print(num_unique_job_titles)
```

```python

insert_at = len(dataset.columns) - 1
dataset.insert(insert_at, 'JobTitle_freq_enc', dataset['Job Title'].map(dataset['Job Title'].value_counts()))
dataset = dataset.drop(columns=['Job Title'])
```

```python
dataset
```

```python
X = dataset.iloc[: , :-1,].values
```

```python
y = dataset.iloc[: , -1].values
```

```python
print(X)
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[: , 1] = le.fit_transform(X[: , 1])
```

```python
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[: , 2] = le.fit_transform(X[: , 2])
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state= 0)
```

Random forest regression implementation

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score_model = r2_score(y_test,y_pred)
```

```python
print("r2_ score of model: " , r2_score_model)
```
